```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("iris")
df
```

```python
df.head(2)
```

```python
df["species"].unique()
```

```python
df["species"]=df["species"].map({"setosa":0,"versicolor":1,"virginica":2})
```

```python
df
```

```python
df["species"]=df["species"].replace({0:"zero",1:"one",2:"two"})
df
```

```python
# df["species"].astype("int" or "float") #to change int in float or float in int 
```

```python
pd.get_dummies(df["species"],drop_first= True) #ohe hot encoding (get dummmies )
```

```python
df1=pd.get_dummies(df["species"],drop_first= True)
```

```python
df1
```

```python
df=pd.concat([df,df1],axis=1) 
```

```python
df
```

```python
df["sq_sepal_length"]=df["sepal_length"].apply(lambda x:x**2)
```

```python
df
```

```python
df["species_s"]=df["species"].apply(lambda x:len(x))
df
```

```python
df["species_even"]=df["petal_length"].apply(lambda x:x//2==0)
df
```

```python
df["petal_length"]=df["petal_length"].astype("int")
```

```python
df["odd_even"]=df["petal_length"].apply(lambda x:x%2==0)
df
```
