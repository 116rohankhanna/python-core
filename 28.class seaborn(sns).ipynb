```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style="darkgrid") #for dark background 
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,hue="day");
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,hue="day",size="sex");
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,hue="day",size="sex",aspect=1,height=7); #for increasing heigth of graph or use figure
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,hue="day",size="sex",style="time");
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,hue="size",size="sex",style="time");
```

# violin chart

```python
sns.catplot(x='day',y='total_bill',hue='sex',kind='violin',data=tips);
```

```python

g=sns.catplot(x='day',y='tip',kind='violin',inner=None,data=tips);
sns.swarmplot(x='day',y='tip',color='k',size=3,data=tips,ax=g.ax);
```
